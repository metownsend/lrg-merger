```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import csv
import numpy as np
import pandas as pd
```

```python
# For gflux and rflux values of every galaxy source in SDSS
# Read in the data 

flux = pd.read_csv("https://raw.githubusercontent.com/metownsend/Data/master/sdss_flux.csv", sep=',', header=0)
```

```python
# Assign data to a variable

gflux = flux['gflux'].values
rflux = flux['rflux'].values
```

```python
# Calculate magnitudes and color

gmag = 22.5 - 2.5 * np.log10(gflux[np.where((gflux > 0.) & (rflux > 0.))])
rmag = 22.5 - 2.5 * np.log10(rflux[np.where((gflux > 0.) & (rflux > 0.))])

color = gmag - rmag
```

```python
# CM plot

plt.scatter(rmag, color, s = 1, color = 'blue')
plt.gca().invert_xaxis()

plt.grid(True)
plt.title('Gmag-Rmag vs. Rmag')
plt.xlabel(r'$r magnitude$')
plt.ylabel(r'$g-r magnitude$')

plt.show()

# I mean...at least the slope is vaguely correct?
```

```python
# For gflux and rflux values of every galaxy source in SDSS in the DECaLS EDR footprint
# Read in the data 

flux1 = pd.read_csv("https://raw.githubusercontent.com/metownsend/Data/master/sdss_EDR.csv", sep=',', header=0)
```

```python
# Assign data to a variable

gflux1 = flux1['gflux'].values
rflux1 = flux1['rflux'].values
```

```python
# Calculate magnitudes and color

gmag1 = 22.5 - 2.5 * np.log10(gflux1[np.where((gflux1 > 0.) & (rflux1 > 0.))])
rmag1 = 22.5 - 2.5 * np.log10(rflux1[np.where((gflux1 > 0.) & (rflux1 > 0.))])

color1 = gmag1 - rmag1
```

```python
# CM plot

plt.scatter(rmag1, color1, s = 1, color = 'purple')
plt.gca().invert_xaxis()

plt.grid(True)
plt.title('Gmag-Rmag vs. Rmag')
plt.xlabel(r'$r magnitude$')
plt.ylabel(r'$g-r magnitude$')

plt.show()

# Vaguely correct slope again?
```
